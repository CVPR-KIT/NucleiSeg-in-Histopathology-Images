```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
label0 = Inslabel[0]
print(np.unique(label0))
```

```python
def correctLabel(label):
    # change all non black pixel to white
    for i in range(0,label.shape[0]):
        for j in range(0,label.shape[1]):
            if label[i][j] != 0:
                label[i][j] = 255
    return label * 255
```

```python
plt.imshow(correctLabel(label0), cmap='gray')
```
